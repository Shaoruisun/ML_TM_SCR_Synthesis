```python
pattern = re.compile(r'(?i)(selectivity|activity|conversion)[^.!?]*?(\d+\.?\d*)')
```

```python
import re
folder_path = '/Users/lsy/work/1-work/5--期刊邀稿/absrtact_2305'
conversion_values = {}
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path,'r') as f:
            text = f.read()
            #text = cleaned_text(text)
            #print(text)

        # 以英文句号为标志分句
        sentences = re.split(r'\. ?', text)
        
        # 初始化一个空列表来存储包含 'conversion' 的句子
        conversion_sentences = []
        
        # 遍历每个句子，检查是否包含 'conversion'，不区分大小写
        for sentence in sentences:
            if re.search(r'conversion', sentence, re.IGNORECASE):
                conversion_sentences.append(sentence.strip())  # 去除前后的空格
        
        # 初始化一个字典来存储每个句子中的数值
        
        print(conversion_sentences)
        # 在列表中遍历每个句子，查找 'conversion' 前后20个字符内的数值
        for sentence in conversion_sentences:
            match = re.search(r'conversion', sentence, re.IGNORECASE)
            if match:
                # 找到 'conversion' 位置
                start = max(match.start() - 30, 0)  # 确保起始位置不小于 0
                end = match.end() + 30  # 终止位置为 'conversion' 后20个字符
                surrounding_text = sentence[start:end]
                
                # 查找前后20个字符内的数值
                numbers = re.findall(r'\d\d.?%|0\.\d{2,4}', surrounding_text)
                if numbers:
                    conversion_values[filename] = numbers
        
        # 输出包含转换率的句子及其对应的数值
        #print(conversion_values)
```

```python
# 将每个字典转换为DataFrame，并用索引（keys）进行拼接
df_conversion = pd.DataFrame(list(conversion_values.items()), columns=['article', 'Conversion'])
df_selectivity = pd.DataFrame(list(selectivity_values.items()), columns=['article', 'Selectivity'])
df_activity = pd.DataFrame(list(activity_values.items()), columns=['article', 'Activity'])

# 合并所有数据框，按 'Sample' 字段进行合并，使用外连接来保留所有样品
df_final = pd.merge(df_conversion, df_selectivity, on='article', how='outer')
df_final = pd.merge(df_final, df_activity, on='article', how='outer')

# 将缺失值填充为空或0（根据需求选择）
df_final.fillna('', inplace=True)

# 保存到Excel文件
file_path = "abstract_conversion_selectivity_activity_results.xlsx"
df_final.to_excel(file_path, index=False)

print(f"数据已保存到 {file_path}")
```
