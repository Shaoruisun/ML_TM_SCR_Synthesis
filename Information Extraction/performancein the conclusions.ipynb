```python
import os
import re
import pandas as pd

```

```python
#\d\d>?%|0\.\d{2,4}
# 定义正则表达式，用于匹配带有Selectivity, Activity, Conversion的句子，并提取数值
folder_path = '/conclusion_2305' 
pattern = re.compile(r'(?i)(selectivity|activity|conversion)[^.!?]*?(\d+\.?\d*)')
```

```python

conversion_values = {}
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path,'r') as f:
            text = f.read()
            #text = cleaned_text(text)
            #print(text)
        sentences = re.split(r'\. ?', text)
        
        conversion_sentences = []
    
        for sentence in sentences:
            if re.search(r'conversion', sentence, re.IGNORECASE):
                conversion_sentences.append(sentence.strip())  
        
        print(conversion_sentences)
    
        for sentence in conversion_sentences:
            match = re.search(r'conversion', sentence, re.IGNORECASE)
            if match:
                
                start = max(match.start() - 20, 0)  
                end = match.end() + 20  
                surrounding_text = sentence[start:end]
                
                numbers = re.findall(r'\d\d.?%|0\.\d{2,4}', surrounding_text)
                if numbers:
                    conversion_values[filename] = numbers
        
        # 输出包含转换率的句子及其对应的数值
        #print(conversion_values)
```

```python
import re
selectivity_values = {}
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path,'r') as f:
            text = f.read()
            #text = cleaned_text(text)
            #print(text)

        sentences = re.split(r'\. ?', text)
        
      
        selectivity_sentences = []
    
        for sentence in sentences:
            if re.search(r'selectivity', sentence, re.IGNORECASE):
                selectivity_sentences.append(sentence.strip())  

        
        print(selectivity_sentences)
        for sentence in selectivity_sentences:
            match = re.search(r'selectivity', sentence, re.IGNORECASE)
            if match:
                start = max(match.start() - 20, 0)  
                end = match.end() + 20  
                surrounding_text = sentence[start:end]
    
                numbers = re.findall(r'\d\d.?%|0\.\d{2,4}', surrounding_text)
                if numbers:
                    selectivity_values[filename] = numbers
        
        # 输出包含转换率的句子及其对应的数值
        #print(selectivity_values)
```

```python
import re
activity_values = {}
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        
        with open(file_path,'r') as f:
            text = f.read()
            #text = cleaned_text(text)
            #print(text)
        sentences = re.split(r'\. ?', text)

        Activity_sentences = []
        for sentence in sentences:
            if re.search(r'Activity', sentence, re.IGNORECASE):
                Activity_sentences.append(sentence.strip())  
        
        
        print(Activity_sentences)

        for sentence in Activity_sentences:
            match = re.search(r'Activity', sentence, re.IGNORECASE)
            if match:
                # 找到 'Activity' 位置
                start = max(match.start() - 20, 0)  
                end = match.end() + 20 
                surrounding_text = sentence[start:end]
                
        
                numbers = re.findall(r'\d\d.?%|0\.\d{2,4}', surrounding_text)
                if numbers:
                    activity_values[filename] = numbers
        
        # 输出包含转换率的句子及其对应的数值
        #print(Activity_values)
```

```python

df_conversion = pd.DataFrame(list(conversion_values.items()), columns=['article', 'Conversion'])
df_selectivity = pd.DataFrame(list(selectivity_values.items()), columns=['article', 'Selectivity'])
df_activity = pd.DataFrame(list(activity_values.items()), columns=['article', 'Activity'])


df_final = pd.merge(df_conversion, df_selectivity, on='article', how='outer')
df_final = pd.merge(df_final, df_activity, on='article', how='outer')


df_final.fillna('', inplace=True)


file_path = "conclusion_conversion_selectivity_activity_results.xlsx"
df_final.to_excel(file_path, index=False)

print(f"数据已保存到 {file_path}")
```
