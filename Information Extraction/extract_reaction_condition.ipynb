```python
import re
import os
import pandas as pd
```

```python
reaction_folder = '/SCR_reation_446/'#Paragraph on catalyst reactions
```

```python
doi = pd.read_excel('DOI_446.xlsx')
doi
```

```python
GHSV = []
for i in range(len(doi)):
    file_path = reaction_folder+doi.iloc[i,1]+'.txt'
    with open(file_path,'r') as file:
        data = file.read()
        #print(data)

        keywords = ['space velocity', 'GHSV']

        sentences = data.split('.')
       
        #pattern = r'(\d{1,4} ?,?\d{3})\s*(h−1|h⁻¹|s⁻¹|m/s|mL/(h·g)|/h|hr−1)' 
        pattern = r'(\d{1,4} ?,?\d{3},?\d?\d?\d?)\s*(h−1|h⁻¹|s⁻¹|m/s|mL/(h·g)|mL·g−1·h−1|/h|hr−1)' 

       
        found = False
        dic = {}
        for sentence in sentences:
            #print(sentence)
            if any(keyword in sentence for keyword in keywords):
                found = True
                match = re.search(pattern, sentence)
                if match:
                    print(match)
                    value = match.group()
                else:
                    value = 'No values'
            #else:
                #value = 'No GHSV'       
                
    dic['filename'] = doi.iloc[i,0]
    dic['GHSV'] = value
    GHSV.append(dic)
                
        
```

```python
temps = []
for i in range(len(doi)):
    file_path = reaction_folder+doi.iloc[i,1]+'.txt'
    with open(file_path,'r') as file:
        data = file.read()
    sentences = re.split(r'(?<=\.)\s*', data)

    dic = {}
    for sentence in sentences:
        #pattern1 = r"\b(\d+\.?\d*)\s*(°C|K)\b"
        #pattern1 = r"\b(\d+\.?\d*)\s*(–|to)?\s*(\d+\.?\d*)?\s*(°C|K)\b"
        pattern1 = r"\b(\d+\.?\d*)\s*(–|to)?\s*(\d+\.?\d*)?\s*(°C|K)\b(?!\/[a-zA-Z]+)"
        
        

    
        match1 = re.search(pattern1, sentence)
        if match1:
            temp = match1.group()
            print(f"Matched1 sentence: {sentence.strip()}")
            print(f'temperature:{temp}')
            break
        pattern2 = r"\b(ambient|room temperature)\b"
        match2 = re.search(pattern2, sentence, re.IGNORECASE)
        if match2:
            temp = match2.group()
            print(f"Matched2 sentence: {sentence.strip()}")
            print(f'temperature:{temp}')
            break 
    else:
        temp = 'None'
        print("No temperature information found.")
    dic['filename'] = doi.iloc[i,1]+'.txt'
    dic['temperature'] = temp
    temps.append(dic)
    print('----'*20)
```
